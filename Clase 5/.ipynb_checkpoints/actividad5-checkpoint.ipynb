# Comparación de Modelos de Clasificación: Árbol de Decisión vs K-NN

En esta actividad, utilizaremos el conjunto de datos Digits de la biblioteca scikit-learn para implementar y comparar dos algoritmos de clasificación: el Árbol de Decisión y K-Nearest Neighbors (K-NN).


```python
# 1. Importar las librerías necesarias
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Cargar el conjunto de datos
digits = load_digits()
X, y = digits.data, digits.target

# 3. Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 4. Entrenar el modelo de Árbol de Decisión
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# 5. Predecir con el modelo de Árbol de Decisión
y_pred_dt = dt_model.predict(X_test)

# 6. Evaluar el modelo de Árbol de Decisión
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print(f"Precisión del Árbol de Decisión: {accuracy_dt:.2f}")
print("Reporte de clasificación para Árbol de Decisión:")
print(classification_report(y_test, y_pred_dt))

# 7. Visualizar la matriz de confusión para Árbol de Decisión
cm_dt = confusion_matrix(y_test, y_pred_dt)
plt.figure(figsize=(10, 8))
sns.heatmap(cm_dt, annot=True, fmt="d", cmap="Blues", 
            annot_kws={"size": 16, "fontweight": 'bold', "color": "white"}, 
            linewidths=0.5, linecolor='gray')
plt.xlabel('Predicciones')
plt.ylabel('Valores Reales')
plt.title('Matriz de Confusión - Árbol de Decisión')
plt.tight_layout()
plt.show()
```

```python
# 8. Entrenar el modelo de K-NN
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)

# 9. Predecir con el modelo de K-NN
y_pred_knn = knn_model.predict(X_test)

# 10. Evaluar el modelo de K-NN
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(f"Precisión del K-NN: {accuracy_knn:.2f}")
print("Reporte de clasificación para K-NN:")
print(classification_report(y_test, y_pred_knn))

# 11. Visualizar la matriz de confusión para K-NN
cm_knn = confusion_matrix(y_test, y_pred_knn)
plt.figure(figsize=(10, 8))
sns.heatmap(cm_knn, annot=True, fmt="d", cmap="Blues", 
            annot_kws={"size": 16, "fontweight": 'bold', "color": "white"}, 
            linewidths=0.5, linecolor='gray')
plt.xlabel('Predicciones')
plt.ylabel('Valores Reales')
plt.title('Matriz de Confusión - K-NN')
plt.tight_layout()
plt.show()



```

```python
# 12. Comparación final
print("\nComparación de modelos:")
if accuracy_dt > accuracy_knn:
    print("El Árbol de Decisión tiene mejor precisión.")
elif accuracy_dt < accuracy_knn:
    print("El K-NN tiene mejor precisión.")
else:
    print("Ambos modelos tienen la misma precisión.")

```
